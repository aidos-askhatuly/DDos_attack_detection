## 1. Load the Data <a name="data_3"></a>

```python
import pandas as pd
```

```python
# specify the path for a parquet file
file_path = 'dataset/network_df.parquet'
```

```python
# read the parquet file
df = pd.read_parquet(file_path, engine = 'pyarrow')
```

## 2. Split the Dataset Into Train and Test <a name="split_2"></a>

```python
X=df.drop(columns='PKT_CLASS')
y=df['PKT_CLASS']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=34, stratify=y)
```

## 3. Balance The Target Class With Smote <a name="smote_2"></a>

```python
from imblearn.over_sampling import SMOTE

# Define the SMOTE
smote = SMOTE(random_state=42)

# Fit and apply the undersampler to the training data
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
# Now X_train_resampled and y_train_resampled contain the balanced dataset
```

```python
X_train.shape # initial training dataset
```

```python
X_train_smote.shape # oversampled training dataset
```

```python
y_train_smote.shape # oversampled target value
```

```python
y_train_smote.value_counts() #number of positive and negative classes is equal after oversampling
```

The training set was oversampled through increasing data points of positive class.

## 4. Feature Scaling <a name="scaling_2"></a>

```python
from sklearn.preprocessing import StandardScaler


# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(X_train_smote)

# Transform both the training and testing data using the fitted scaler
X_train_scaled = scaler.transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)  # Use the same scaler fitted on X_train
```

## 5. Models Optimization <a name="optimization"></a>

### Logistic Regression

To find the best hyperparameters, we will experiement with `regularization` (C value) and `penalty type` (L1 and L2).

### Penalty

```python
from tempfile import mkdtemp
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
# Set up a directory to cache the pipeline results - if the same step with the same parameters is called again, the cached result is used instead of re-computing it.
cachedir = mkdtemp()

# Set up a pipeline
# The steps here act as placeholders and will be changed when we pass the pipeline into the grid search later
my_pipeline = Pipeline(steps=[
                        ('model', LogisticRegression())],
                        memory=cachedir) # Set up the caching
```

```python
# Let's try the same range of C values from earlier
c_values = [.00001, .0001, .001, .1, 1, 10, 100, 1000, 10000]

# Parameter grid
logreg_param_grid = [

    # l1
    {
     'model': [LogisticRegression(solver='saga',penalty='l1', random_state=1, n_jobs=-1, max_iter=100000)],
     'model__C': c_values},


    # l2
    {
     'model': [LogisticRegression(penalty='l2',random_state=1, n_jobs=-1, max_iter=100000)],
     'model__C': c_values}
]
```

```python
# training the model
import time
start = time.time()
# Instantiate the log reg grid search
logreg_gs = GridSearchCV(my_pipeline, param_grid=logreg_param_grid, cv=5, n_jobs=-1, verbose=1, scoring='f1', refit=True)

# Fit the log reg grid search
logreg_gs.fit(X_train_scaled, y_train_smote)

end=time.time()
print(end-start, "seconds")
```

```python
# Retrieve best parameters and score
print("Best Parameters:", logreg_gs.best_params_)
print("Best Score:", logreg_gs.best_score_)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score,roc_auc_score

# Predict on the training set
y_pred_train_logreg = logreg_gs.predict(X_train_scaled)

# Calculate accuracy, recall, f1 and roc auc score
accuracy_train = accuracy_score(y_train_smote, y_pred_train_logreg)
recall_train = recall_score(y_train_smote, y_pred_train_logreg)
f1_train = f1_score(y_train_smote, y_pred_train_logreg)
roc_auc_train = roc_auc_score(y_train_smote, y_pred_train_logreg)

print("Train Accuracy:", accuracy_train)
print("Train Recall:", recall_train)
print("Train F1-score:", f1_train)
print("Train ROC AUC Score:", roc_auc_train)
```

```python
# Get the best estimator from the grid search
best_estimator = logreg_gs.best_estimator_

# Predict on the test set
y_pred_logreg = best_estimator.predict(X_test_scaled)

# Calculate evaluation metrics
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
recall_logreg = recall_score(y_test, y_pred_logreg)
f1_logreg = f1_score(y_test, y_pred_logreg)
roc_auc_logreg = roc_auc_score(y_test, y_pred_logreg)

print("Test Accuracy:", accuracy_logreg)
print("Test Recall:", recall_logreg)
print("Test F1:", f1_logreg)
print("Test ROC AUC:", roc_auc_logreg)
```

### Dense Network Hyperparameter Tuning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import backend as K
import time
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
import multiprocessing
```

#### Number of Layers

```python
def build_model(num_layers):
    """
    Builds and compiles a neural network model with the specified number of hidden layers.

    Parameters:
    num_layers (int): The number of hidden layers to include in the model.

    Returns:
    model: A compiled Keras Sequential model.
    """
    # Initialize a Sequential model
    model = models.Sequential()
    
    # Add an input layer with the shape of the training data
    model.add(layers.InputLayer(input_shape=(X_train_scaled.shape[1],)))

    # Add the specified number of hidden layers with 16 neurons each and 'tanh' activation function
    for _ in range(num_layers):
        model.add(layers.Dense(16, activation='tanh'))
    
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=keras.metrics.Recall())
    return model
```

```python
#Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss',  # Monitor validation loss
                               patience=3,         # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True)  # Restore model weights from the epoch with the best value of the monitored metric
```

```python
# Parameters
num_layers_list = [1,2,3,4,5] # List of different numbers of layers to experiment with
epochs = 50 # Number of epochs for training each model

# Dictionary to store the results of each model
results = {}
results_f1 = {}
start = time.time()

# Loop through each number of layers specified in num_layers_list
for num_layers in num_layers_list:
    print(f'Training model with {num_layers} layers')
    
    # Build a model with the specified number of layers
    model = build_model(num_layers)
    
    # Train the model with the specified parameters
    history = model.fit(X_train_scaled,  # Scaled training data
                        y_train_smote,  # Training labels with SMOTE applied
                        epochs=epochs, # Number of epochs
                        validation_data=(X_test_scaled, y_test),  # Validation data and labels
                        verbose=1,
                        callbacks=[early_stopping], # Early stopping callback to prevent overfitting
                        workers=multiprocessing.cpu_count(), # Use all available CPU cores for parallel processing
                        use_multiprocessing=True) # Enable multiprocessing
   
    # Store the history of training results for the current model
    results[num_layers] = history.history
    
    # Make predictions on the test set
    y_pred_dense = model.predict(X_test_scaled)

    # Since the output is in probabilities, convert them to binary predictions
    y_pred_binary = (y_pred_dense > 0.5).astype(int)

    # Calculate the F1 score
    f1 = f1_score(y_test, y_pred_binary)

    results_f1[num_layers] = f1
end = time.time()

print(end-start)
```

#### Number of Nodes

```python
def build_model(num_nodes):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train.shape[1],)))
    
    # Keeping the number of layers constant as 3
    for _ in range(3):
        model.add(layers.Dense(num_nodes, activation='tanh'))
    
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=keras.metrics.Recall())

    return model
```

```python
#Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss',  # Monitor validation loss
                               patience=3,         # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True)  # Restore model weights from the epoch with the best value of the monitored metric
```

```python
# Parameters
num_nodes_list = [4, 8, 16, 32, 64]  # Different configurations for the number of nodes
epochs = 50

# Store results
results = {}
start = time.time()
for num_nodes in num_nodes_list:
    print(f'Training model with {num_nodes} nodes per layer')
    model = build_model(num_nodes)
    history = model.fit(X_train_scaled, 
                        y_train_smote, 
                        epochs=epochs, 
                        validation_data=(X_test_scaled, y_test), 
                        verbose=1,
                        callbacks=[early_stopping], #early stopping
                        workers=multiprocessing.cpu_count(), 
                        use_multiprocessing=True)
    results[num_nodes] = history.history

end = time.time()
print(end-start)
```

#### Batch Size

Keras uses default batch size 32

```python
def build_model():
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train.shape[1],)))
    
    # Adding three layers with a fixed number (4) of nodes
    for _ in range(3):
        model.add(layers.Dense(4, activation='tanh'))
    
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=keras.metrics.Recall())
    return model
```

```python
# Parameters
batch_size_list = [16, 32, 64, 128]
epochs = 50
# Store results
results = {}
```

```python
#Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss',  # Monitor validation loss
                               patience=3,         # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True)  # Restore model weights from the epoch with the best value of the monitored metric
```

```python
for batch_size in batch_size_list:
    print(f'Training model with batch size {batch_size}')
    model = build_model()
    history = model.fit(
        X_train_scaled, y_train_smote,
        epochs=epochs,
        validation_data=(X_test_scaled, y_test),
        batch_size=batch_size,
        workers=multiprocessing.cpu_count(),
        use_multiprocessing=True,
        verbose=1,
        callbacks=[early_stopping], #early stopping
    )
    results[batch_size] = history.history
```

```python
# Plot the results
plt.figure(figsize=(14, 7))

number = 17
for batch_size in batch_size_list:
    metric_name = f'recall_{number}'
    plt.plot(results[batch_size][metric_name], label=f'Batch size {batch_size}')
    number+=1
plt.title('Train Recall vs. Batch Size')
plt.xlabel('Epochs')
plt.ylabel('Train Recall')
plt.legend()
plt.show()
```

#### Dropout Rate

```python
# Define the build_model function with dropout_rate parameter
def build_model(dropout_rate):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train_scaled.shape[1],)))
    
    # Adding two layers with 4 nodes each and 'tanh' activation
    for _ in range(3):
        model.add(layers.Dense(4, activation='tanh'))
        model.add(layers.Dropout(dropout_rate))  # Add dropout layer with specified rate
    
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=tf.keras.metrics.Recall())
    return model
```

```python
# Parameters
dropout_rate_list = [0.1, 0.3, 0.5, 0.7]
epochs = 50

# Store results
results = {}
```

```python
#Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss',  # Monitor validation loss
                               patience=4,         # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True)  # Restore model weights from the epoch with the best value of the monitored metric
```

```python
 #Train the model with different dropout rates
for dp in dropout_rate_list:
    print(f'Training model with dropout rate {dp}')
    model = build_model(dp)
    history = model.fit(
        X_train_scaled, y_train_smote,
        epochs=epochs,
        validation_data=(X_test_scaled, y_test),
        batch_size=16,  # Adjust batch size as needed
        workers=multiprocessing.cpu_count(),
        use_multiprocessing=True,
        verbose=1,
        callbacks=early_stopping
    )
    results[dp] = history.history
```

```python
# Plotting the recall for each dropout rate
plt.figure(figsize=(14, 7))
number = 31
for dropout_rate in dropout_rate_list:
    metric_name = f'recall_{number}'
    plt.plot(results[dropout_rate][metric_name], label=f'Dropout Rate {dropout_rate}')
    number+=1
    
plt.title('Training Recall vs. Dropout Rate')
plt.xlabel('Epochs')
plt.ylabel('Training Recall')
plt.legend()
plt.show()
```
